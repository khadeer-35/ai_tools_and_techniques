```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd
```

```python
#reading data
data=pd.read_csv("profitdataset (1).txt")
print(data.shape)
```

```python
# get x and y values
x=data['population'].values
y=data['profit'].values
x
```

```python
#plot the data values
plt.scatter(x,y,50,label='profit calculation',c='k',marker='x')
plt.xlabel("population")
plt.ylabel("profit")
plt.title("fodd truck profit calculation")
plt.grid(True,color='y')
plt.show
```

```python
mean_x=np.mean(x)
mean_y=np.mean(y)
n=len(x)
```

```python
#to find m and c
num=0
den=0
for i in range(n):
    num=num+((x[i]-mean_x)*(y[i]-mean_y))
    den=den+((x[i]-mean_x)**2)
m=num/den
#c=y-mx
c=mean_y-(m*mean_x)
print(m,c)
```

```python
# calculate y predicate
y_pred=np.array([])
for i in range(n):
    p=(m*x[i])+c
    y_pred=np.append(y_pred,np.array([p]),axis=0)
print(y_pred)    
```

```python
#plot the regression plot
plt.plot(x,y_pred,color='m',label='regression line')
#plot the actual data
plt.scatter(x,y,40,c='g',label="actual_data",marker='o')
plt.scatter(x,y_pred,40,c='k',label='predicted_data',marker='x')
plt.xlabel("population")
plt.ylabel("profit")
plt.legend()
plt.show()
```

```python
#how good our fit line or regression line is by using r2 method
nu=0
de=0
for i in range(n):
    nu=nu+((y_pred[i]-mean_y)**2)
    de=de+((y[i]-mean_y)**2)
r2=nu/de
r2
```
